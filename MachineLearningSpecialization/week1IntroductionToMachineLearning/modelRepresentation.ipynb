# Numpy:

NumPy (Numerical Python) is the foundational Python library for numerical and scientific computing. It introduces the ndarray, a fast, memory-efficient N-dimensional array that supports vectorized operations, making computations much quicker than standard Python lists. NumPy provides essential tools for array creation, indexing, slicing, reshaping, and broadcasting, plus mathematical functions for algebra, statistics, and linear algebra. It also includes random number generation and utilities for working with data types and memory layout. Because of its speed and clean syntax, NumPy is the backbone of many libraries such as SciPy, pandas, and scikit-learn for data science.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(r"B:\22daysChallengeDataScience\MachineLearningSpecialization\week1IntroductionToMachineLearning\deeplearning.mplstyle")
```

```python
# x-trin is the input feature variables.
# y_train is the target varible.

x_train = np.array([1.0 ,2.0]) #Features
y_train = np.array([300.0 , 500.0]) # Target
```

```python
# To print the value of the input features and target.
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# m is the number of the training examples
print(f"x_train.shape: {x_train.shape}") #  shows number of elements exits in the array.
m = x_train.shape[0]
print(f"Number of the training example is: {m}")
```

```python
# m is the number of the training example.
m = len(x_train)
print(f"Number of the training example is : {m}")
```

```python
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"x^({i}) , y^({i}) = ({x_i} , {y_i})") # Power of x and y.
```

```python
#Plot the data points
plt.scatter(x_train , y_train , marker='*' , c='r')

#set the titles
plt.title("House price prediction:")
#set the y-axix label
plt.ylabel('price(in thousands dollars)')

# set the x-axis label
plt.xlabel('size (in 1000sqft)')
plt.show()
```

```python
# initially put the values of the w and b
w = 100
b = 100
print(f'w: {w}')
print(f"b: {b}")
```

```python
#compute the model

def compute_model_output(x , w ,b):
    m = x_train.shape[0]
    f_wb = np.zeros(m) # to store the values of the function in each iteration.
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
temp_f_wb = compute_model_output(x_train , w , b)

plt.plot(x_train , temp_f_wb ,c ='g' , label = 'our prediction')
plt.scatter(x_train , y_train , c= 'r' , marker='x' , label = 'Actual prediction')
plt.title("House price prediction")
plt.ylabel('Price in (1000 of dollars)')
plt.xlabel('Size in (1000sqft)')
plt.legend() # To print the labels of the plot as well
plt.show()
```
